#### Research Question
How does the use of religious rhetoric at the presidential nominating conventions differ across party lines and over time?

#### Hypotheses
Given the existing qualitative literature analyzing religious rhetoric in politics, it can be hypothesized that appeals to religion by Republicans will be more often and explicit. It is expected that Republicans will use rhetoric in a predominantly monotheistic, Protestant Christian context with more direct references to their own religiosity. On the other hand, Democrat’s use of religious rhetoric is likely to be less denominations, engaging rhetoric more rooted in faith generally in a more diverse context. Specifically, it expected of Democrats to make calls to support individuals of all religions. In turn, there will be fewer specific calls to Christian doctrine, instead including more unifying language surrounding religious rhetoric, though likely still monotheistic in nature.


#### Methodology

The analysis here primarily employs transcripts of nominating convention speeches sourced from the University of California, Santa Barbara’s American Presidency Project ([UCSB APP](https://www.presidency.ucsb.edu/)). For this study, the transcripts were web scraped from the American Presidency Project webpage by pulling the document URLs from the html code of each page of speeches under the Convention Speeches section of the Documents page. Using these URLS, the speech transcript was scraped alongside the speaker’s name and the date of the speech. In total, there are 58 speech transcripts spanning both the Republican and Democratic National Conventions from 1932 to 2024. The number of speeches per convention increases with time, the most coming from 2024. The speeches include speeches from the presidential nominee, their vice president, and other prominent key note speakers.

To decide what qualifies as key religious terms, a variety of generative AI platforms, like ChatGPT, Claude, and Gemini, were asked to compile a list of religious terms which were frequently used in political speeches. Using the results from the query, the study focused in on the following fourteen words to conduct the analysis:


*faith, grace, spirit, virtue, redemption, salvation, providence, prayer, bless, heaven, hell, creed, prophet, and god*.

The analysis begins with a glance at concordance and similarity results. At first, this is done with no distinction by political party, building a foundation of the general contexts in which the religious terms are used. For a more in-depth analysis, the text is split into two blocks, each containing the entire body of text of all speeches at a respective Democratic or Republican Convention. This step allows for a cursory analysis at how the context in which the key religious terms are used differs across party lines. From here, the similarity analysis is utilized to examine which other terms in the speeches are used in a similar context to help assess the more qualitative hypotheses described above.

To begin the more quantitative portion of analysis, a frequency count of each key religious word in each convention speech is calculated. To get the most accurate and consistent estimate for each speech, the text is lemmatized and tokenized before counting. This allows for word roots to be emphasized; any past tense and pluralized versions of the above words are thus still included in the count. For the total frequency analysis, the total number of designated religious terms in each text is summed. The data is then collapsed by year and party, creating an normalized average count of religious terms per speech for each party and presidential election year.

Similarly, to evaluate the importance of the listed religious terms in each speech, the term frequency-inverse document frequency will be calculated for each term using the sklearn package in Python. Specifically, the frequency at which a word appears in a certain speech is multiplied by the log of the inverse of the fraction of total speeches which contain that word plus one. Thus, if-idf will assign higher values to the religious terms in a certain speech which are used less frequently in other speeches. From here, the tf-idf scores for each religious word in each speech are summed. For the assessment of religious term importance, the mean total tf-idf score of religious terms for each party for each presidential election year are compared to understand how the utilization of religious terminology has evolved across time for both Republicans and Democrats.

#### Results & Discussion


*I.   Similarity and Concordance Analysis*

When conducting the qualitative analysis using nltk.Text.similar(), there are a few primary differences between the use of religious rhetoric by Democrats and Republicans. First, when using the word ‘god,’ speakers at the Republican National Convention tended to use mentions of God alongside references to their candidates. The words used most frequently in the same context as the word ‘god’ in RNC speeches include ‘me’ and ‘president.’ By employing words which reference the identity of the candidate in similar ways, Republicans are using ‘god’ to refer to the holiness of their candidate. For example, when accepting his nomination at the 2024 RNC, Donald Trump stated “I felt very safe because I had God on my side,” employing rhetoric which positions him as a solider of the divine. On the other hand, usage of the word ‘god’ at the DNC came in more diverse contexts. Among the most frequently used words in a similar context by Democrats is more unifying language, like “all” and “everyone.” By using ‘god’ in this context, while still imposing religiosity onto the nation, speakers at the DNC were less focused on proving themselves to be religious. Instead, they used this rhetoric to paint an image of a one, divinely protected, America.  This affirms the expressed hypothesis which posited an expected diversity centered approach to using religious rhetoric.

Another example of the difference in utilization of religious rhetoric is reflected by the contexts of the word ‘prayer.’ The words used in similar contexts to ‘prayer’ at the RNC seem to emphasize tradition and hardship, specifically ‘history,’ and ‘struggle.’ This likely reflects a view by Republicans of prayer as a tool in times of hardship, while still emphasizing the need for a historic, traditional approach to religion. The centering of tradition is a hallmark of conservative religious ideology. When looking to the use of ‘prayer’ by Democrats, however, the focus is primarily forward looking. The use of words like ‘hope’ and ‘opportunity’ in similar contexts to prayer signifies the tendance of DNC speakers to use prayer as an aspiration or mechanism of improvement. Also consistent with liberal ideology, the emphasis on the future is indicative of the importance of innovation and progress in their platform.

*II.   Frequency Analysis*

Now, on the quantitative side, while the use of religious rhetoric is consistent with the ideological dogma of each party, the frequency and importance of religious terms by political party is less straightforward. First, frequency. While it was expected for RNC speakers to have consistently and increasingly more frequent religious rhetoric than Democrats, this is not the case. While in most years Republicans employed more religious rhetoric than Democrats, there are two trends which stick out. First, in Figure 1, from 1948 to 1972, there is an alternating cycle of peaks of religious frequency between parties. This could be because the parties are adapting to the behavior of the other during this period, manifesting in peaks which mimic each other with a four-to-eight-year lag.

Additionally, it could be hypothesized that this trend manifests due to electoral success; for example, the party who loses the election may be more likely to use religious rhetoric, implying a hypothesized link between using religious rhetoric and winning the election. Second, there is a dramatic drop and increase in frequency of religious rhetoric at the RNC, falling in 2016 and peaking at an all-time, all party high in 2020. This comes during the first of Trump’s three nominating conventions. The dip in frequency of religious rhetoric in his first term could be anticipated due to his platform as a businessman, but the peak sharp jump up in 2020 is fascinating. Perhaps due to the salience of the COVID-19 Pandemic, where many may have turned to religion for hope and comfort, Republican politicians could have sensed this shift in their voter base and responded accordingly.

*II.   TF-IDF Term Importance*
In addition to frequency of religious terms, an investigation into the importance of religious terms at each RNC and DNC since 1932 can elucidate how each party uses religious rhetoric to achieve its goals. Similarly, here it was expected that religious words would hold higher importance at Republican National Conventions than at the Democratic National Conventions. Figure 2 elucidates, from 1960 through 2000, this hypothesis is majorly correct, with the importance of religious terms at the RNC being equal to or higher than the DNC for the entire period, except for 1968. The stage for Nixon’s electoral success, the 1968 RNC had historically low importance of religious terms, likely due to Nixon’s denominational identity as a Quaker ([Quaker.org](https://quaker.org/legacy/athensga/history.html)) Given the insurgence of liberalism in the Quaker church during the civil rights movement, it is likely Nixon attempted to distance himself from his religious identity as he ran on a conservative platform ([Smith](https://academic.oup.com/book/26109/chapter/194106900)).

Similarly to religious frequency, there is also a stark contrast between the use of religion in Trump’s first and second presidential campaigns. In 2016, religious rhetoric had an all-time low importance at the RNC, lower than the DNC for Hillary Clinton’s campaign. However, in 2020, the term importance of religious rhetoric skyrocketed. Given the political climate of that election cycle, with elevated social unrest and the Covid-19 pandemic, republicans and Trump likely leaned into religious themes. Specially, there was debates in evangelical circles, a historically conservative voting bloc, regarding religious freedom amidst forced church closures during the pandemic ([Gjelten](https://www.npr.org/sections/coronavirus-live-updates/2020/04/17/837698597/opposing-forced-church-closures-becomes-new-religious-freedom-cause)).
```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

speech_links = []

for i in range(6):
  i = str(i)
  if i == "0":
    url = "https://www.presidency.ucsb.edu/documents/app-categories/elections-and-transitions/convention-speeches"
  else:
    url = "https://www.presidency.ucsb.edu/documents/app-categories/elections-and-transitions/convention-speeches?page="+i
  print(url)
  response = requests.get(url)
  soup = BeautifulSoup(response.content, "html.parser")
  for link in soup.find_all("a", href=True):
    href = link['href']
    if "/documents/" in href and href.count("/") == 2:
        speech_links.append("https://www.presidency.ucsb.edu" + href)

```

```python
speech_links = list(set(speech_links))

nonspeeches = [ 'https://www.presidency.ucsb.edu/documents/presidential-documents-archive-guidebook', 'https://www.presidency.ucsb.edu/documents/category-attributes']

for i in nonspeeches:
  speech_links.remove(i)


speech_links
```

```python
speeches = []
for speech_url in speech_links:
    speech_response = requests.get(speech_url)
    speech_soup = BeautifulSoup(speech_response.content, "html.parser")

    title = speech_soup.find("div", class_="field-ds-doc-title").get_text().strip()
    speaker = speech_soup.find("div", class_="field-title").get_text().strip()
    date = speech_soup.find("div", class_="field-docs-start-date-time").get_text().strip()
    content = speech_soup.find("div", class_="field-docs-content").get_text().strip()

    speeches.append({
        "Title": title,
        "Speaker": speaker,
        "Date": date,
        "Content": content,
        "URL": speech_url
    })

speeches_df = pd.DataFrame(speeches)
speeches_df.to_csv("convention_speeches.csv", index=False)

```

```python
def clean(text):
    cleaned_text = text.replace("\'", "")
    cleaned_text = cleaned_text.replace("\n", " ")
    return cleaned_text


speeches_df['Content'] = speeches_df['Content'].apply(clean)

speeches_df['Date'] = pd.to_datetime(speeches_df['Date'])

speeches_df = speeches_df.sort_values(by="Date", ascending=False).reset_index()

speeches_df
```

```python
for i in range(len(speeches_df)):
    title = speeches_df.loc[i, 'Title']
    if "RNC" in title or "Republican" in title:
        speeches_df.loc[i, 'Party'] = "Republican"
    if "DNC" in title or "Democratic" in title:
        speeches_df.loc[i, 'Party'] = "Democrat"

speeches_df['Party'] = speeches_df['Party'].fillna("Democrat")

speeches_df.reset_index(drop=True, inplace=True)


speeches_df
```

```python
def count_word_frequencies(df, word_list):
    frequency_list = []

    for i in range(len(df)):
        transcript = df.loc[i, 'Content']
        transcript = transcript.lower()
        word_frequencies = []

        for word in word_list:
            count = transcript.split().count(word.lower())
            word_frequencies.append(count)

        frequency_list.append(word_frequencies)

    df['Frequency'] = frequency_list

    return df

word_list = ['faith', 'grace', 'spirit', 'virtue', 'redemption', 'salvation', 'providence', 'prayer', 'bless', 'heaven', 'hell', 'creed', 'prophet', 'god']
speeches_df = count_word_frequencies(speeches_df, word_list)

speeches_df
```

```python
speeches_df['Name'] = ""

for i in range(len(speeches_df)):
    speeches_df['Name'][i] = str(speeches_df['Date'][i].year)+ "_" + speeches_df['Speaker'][i]

speeches_df
```

```python
import nltk

nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('punkt_tab')
nltk.download('stopwords')

from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk import pos_tag
from nltk.corpus import stopwords
stops = stopwords.words('english')
```

```python
lemmatizer = WordNetLemmatizer()

def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = pos_tag([word])[0][1][0].upper()  # Get the first letter of the POS tag
    tag_dict = {
        'J': wordnet.ADJ,  # Adjective
        'N': wordnet.NOUN,  # Noun
        'V': wordnet.VERB,  # Verb
        'R': wordnet.ADV    # Adverb
    }
    return tag_dict.get(tag, wordnet.NOUN)  # Default to noun if not found
```

```python

dem_text = ""

for i in speeches_df.iterrows():
  if i[1]['Party'] == "Democrat":
    dem_text = dem_text + i[1]['Content']

dem_text = dem_text.lower()

dem_text = nltk.Text(nltk.word_tokenize(dem_text))

for word in word_list:
    print(f" The following are examples of the usage of {word} in DNC Speeches: ")
    dem_text.concordance(word)

    print(f" The following are words often used in similar context to {word} in DNC Speeches: ")
    dem_text.similar(word)
```

```python
rep_text = ""

for i in speeches_df.iterrows():
  if i[1]['Party'] == "Republican":
    rep_text = rep_text + i[1]['Content']

rep_text = rep_text.lower()

rep_text = nltk.Text(nltk.word_tokenize(rep_text))

for word in word_list:
    print(f" The following are examples of the usage of {word} in RNC Speeches: ")
    rep_text.concordance(word)

    print(f" The following are words often used in similar context to {word} in RNC Speeches: ")
    rep_text.similar(word)
```

```python
## Generated by ChatGPT "Create a df which shows the words from .silimar()
##                       which are different for each party"

import contextlib
import io
# Creating DataFrame
rows = []

for word in word_list:
    # Capture similar words for democratic text
    with contextlib.redirect_stdout(io.StringIO()) as f:
        dem_text.similar(word)
    dem_similar_words = f.getvalue().strip().split('\n')[0].split()

    # Capture similar words for republican text
    with contextlib.redirect_stdout(io.StringIO()) as f:
        rep_text.similar(word)
    rep_similar_words = f.getvalue().strip().split('\n')[0].split()

    # Calculate common and unique words
    common_words = list(set(dem_similar_words).intersection(rep_similar_words))
    unique_dem = list(set(dem_similar_words).difference(rep_similar_words))
    unique_rep = list(set(rep_similar_words).difference(dem_similar_words))

    # Append data to the rows list
    rows.append({
        'Word': word,
        'Dem_Similar': dem_similar_words,
        'Rep_Similar': rep_similar_words,
        'Common': common_words,
        'Unique_Dem': unique_dem,
        'Unique_Rep': unique_rep
    })

# Convert the rows list to a DataFrame
similar_df = pd.DataFrame(rows)
```

```python
def count_total_word_frequencies(df, word_list):
    lemmatizer = WordNetLemmatizer()
    frequency_list = []
    for idx, row in df.iterrows():
        transcript = row['Content']
        transcript = transcript.lower()

        words = nltk.word_tokenize(transcript)
        pos_tags = nltk.pos_tag(words)
        lemmatized_words = [
            lemmatizer.lemmatize(word, get_wordnet_pos(tag))
            for word, tag in pos_tags
        ]

        total_count = 0

        for word in word_list:
            lemmatized_word = lemmatizer.lemmatize(word.lower())
            total_count += lemmatized_words.count(lemmatized_word)

        frequency_list.append(total_count)

    df['Total_Frequency'] = frequency_list

    return df

speeches_df = count_total_word_frequencies(speeches_df, word_list)
speeches_df
```

```python
speeches_df['Year'] = speeches_df['Date'].dt.year

grouped_freq_stats = speeches_df.groupby(['Year', 'Party']).agg(
    avg_freqency=('Total_Frequency', 'mean'),
).reset_index()

grouped_freq_stats
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 6))

for party in grouped_freq_stats['Party'].unique():
    party_data = grouped_freq_stats[grouped_freq_stats['Party'] == party]

    if party == 'Republican':
        plt.plot(party_data['Year'], party_data['avg_freqency'], label=party, color='red', lw=2)
    else:
        plt.plot(party_data['Year'], party_data['avg_freqency'], label=party, lw=2)

plt.xticks(
    ticks=range(int(grouped_freq_stats['Year'].min()),
                int(grouped_freq_stats['Year'].max()) + 1, 4),
    rotation=45)
plt.xlabel('Year')
plt.ylabel('Frequency of Religious Terms')
plt.title('Figure 1: Frequency of Religious Terms Over Time by Party')
plt.legend()
plt.grid(True)

```

```python
speeches_df['Lemmetized'] = ""


for i in range(len(speeches_df)):
  text = speeches_df.loc[i, 'Content']
  text_tokens = nltk.word_tokenize(text)
  nltk_text = nltk.Text(text_tokens)
  text_lower = [t.lower() for t in nltk_text if t.isalnum()]
  text_stops = [t for t in text_lower if t not in stops]
  text_clean = [WordNetLemmatizer().lemmatize(t, get_wordnet_pos(t)) for t in text_stops]
  text_string = ' '.join(text_clean)
  speeches_df.loc[i, 'Lemmetized'] = text_string
```
